```
from gensim.models import Word2Vec
story=open("jt.txt","r")
story=story.read()
#print(story)
```

```
import re
data=re.sub(r'\[(0-9)*\]',' ',story)
data=re.sub(r'[\s+]',' ',data)
data=data.lower()
data=re.sub(r'[\d]',' ',data)
data=re.sub(r'[,]',' ',data)
data=re.sub(r'[\s+]',' ',data)
print(data)
```

```
import nltk
nltk.download('punkt')
from nltk.tokenize import sent_tokenize,word_tokenize
```

```
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words=set(stopwords.words("english"))
```

```
sentences=nltk.sent_tokenize(data)
sentences=[nltk.word_tokenize(sentence) for sentence in sentences]
print(sentences)
```

```
from nltk.stem import PorterStemmer,WordNetLemmatizer
from nltk.corpus import wordnet
nltk.download('wordnet')
for i in range(len(sentences)):
  sentences[i]=[word for word in sentences[i] if word not in stop_words]
  sentences[i]=[WordNetLemmatizer().lemmatize(word,pos=wordnet.VERB) for word in sentences[i]]
print(sentences)
```

```
model1=Word2Vec(sentences,min_count=1)
model1.wv.vocab
```
