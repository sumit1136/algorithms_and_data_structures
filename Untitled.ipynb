```
from nltk.corpus import stopwords
stop_words=set(stopwords.words("english"))
#print(stop_words)
from nltk.tokenize import word_tokenize
words=nltk.word_tokenize(text)
#print(words)
word=[word for word in words if not word in stop_words]
print(word)
```

```
data=re.sub(r'\[(0-9)*\]',' ',data)
data=re.sub(r'\[s+]',' ',data)
data=data.lower()
data=re.sub(r'\[d]',' ',data)
data=re.sub(r'\['r'+]',' ',data)
data=re.sub(r'\[s+]',' ',data)
sentences=nltk.sent_tokenize(data)
sentences=[nltk.word_tokenize(sentence) for sentence in sentences]
#print(sentences)
for i in range(len(sentences)):
  sentences[i]=[word for word in sentences[i] if word not in stop_words]
  sentences[i]=[WordNetLemmatizer().lemmatize(word,pos=wordnet.VERB) for word in sentences[i]]
```

```
model1=Word2Vec(sentences,min_count=1)
word3=model1.wv.vocab
#print(word3)
#print(sentences)
#print(model1.wv['tales'])
similar=model1.wv.most_similar('jataka')
print(similar)
```
